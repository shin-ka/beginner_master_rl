## Import the necessary software libraries:

```python
import numpy as np
import matplotlib.pyplot as plt

from envs import Maze
from utils import plot_policy, plot_action_values, test_agent
```

## Initialize the environment

```python
env = Maze()
```

```python
frame = env.render(mode='rgb_array')
plt.axis('off')
plt.imshow(frame)
```

## Define value table $Q(s, a)$

#### Create the $Q(s, a)$ table

```python
action_values = np.zeros((5,5,4))
```

#### Plot $Q(s, a)$

```python
plot_action_values(action_values)
```

## Define the ϵ-greedy policy $\pi(s)$

#### Create the policy $\pi(s)$

```python
def policy(state, epsilon=0.2):
    p = np.random.random() # selects a number from [0.0,1.0) from a uniform distribution
    if p < epsilon:
        return np.random.choice(4) #select one of the available actions randomly
    else:
        av = action_values[state] 
        # av is vector with four values, a value for each action in this state, and what it will return
        # is the action with the highest Q value.
        # But if there's a tie between the actions with the highest value, then you do this:
        return np.random.choice(np.flatnonzero(av == av.max()))
# av == av.max() returns an array of True and False | av = [1,2,3,3], then (av == av.max) = [False False True True]
# np.flatnonzero() returns the indices of non-zero elements in array | for above eg it'll be [2 3]
# these indices indeed coincide with the actions, since more than 1 action has higher q value, we'll randomly pick one of them
```

#### Test the policy with state (0, 0)

```python
action = policy((0,0))
print(f"Action taken in state (0,0) is : {action}")
```

#### Plot the policy

```python
plot_policy(action_values, frame)
```

## Implement the algorithm


</br></br></br>




<div style="text-align:center">
    Adapted from Barto & Sutton: "Reinforcement Learning: An Introduction".
</div>

```python
def on_policy_mc_control(policy, action_values, episodes, epsilon=0.2, gamma=0.99):
    sa_returns = {} # state-action returns
    
    for episode in range(1, episodes+1):
        state = env.reset()
        done = False
        transitions = [] # store the trajectory taken by agent
        
        while not done:
            action = policy(state, epsilon)
            next_state, reward, done, _ = env.step(action)
            transitions.append([state, action, reward])
            state = next_state
            
        G = 0
        
        for state_t, action_t, reward_t in reversed(transitions):
            G = reward_t + gamma * G
            #print(f"G(S,A) : {sa_returns}")
            if not (state_t, action_t) in sa_returns: # if we don't have an entry in the dict for state_t and action_t, we create it
                sa_returns[(state_t, action_t)] = []
            sa_returns[(state_t, action_t)].append(G) # line 10
            action_values[state_t][action_t] = np.mean(sa_returns[(state_t, action_t)]) 
            #print(f"action_values : {action_values}")
                
```

```python
on_policy_mc_control(policy, action_values, episodes=1000)
```
